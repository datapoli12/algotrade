```python
import yfinance as yf
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import mplfinance as mpf
```

# Récupération des données financières

```python
def get_df(tickers, start_date, end_date, interval):
    df_dict = {}
    for ticker in tickers:
        df = yf.download(ticker, start=start_date, end=end_date, interval=interval)
        df_dict[ticker] = df
    return df_dict

tickers = ['AAPL', 'MSFT', 'SPY', 'IVV', 'AF.PA', '^FCHI']

tickers_names = {'AAPL': 'Apple', 'MSFT': 'Microsoft', 'SPY': 'SP500', 'IVV': 'iShares Core S&P500 ETF', 'AF.PA': 'Air France', '^FCHI': 'CAC40'}


df_dict = get_df(tickers, '2006-01-01', '2023-01-01', '1d')

### Netttoyage des données

def clean_df(df_dict):
    clean_df_dict = {}
    for ticker, df in df_dict.items():
        cleaned_df = df.dropna()
        clean_df_dict[ticker] = cleaned_df
    return clean_df_dict

clean_df_dict = clean_df(df_dict)

# affichage des données

for ticker in tickers:
    print(f"les données de {ticker} sont les suivantes :")
    print(df_dict[ticker].head())
    print("\n")

    ############

    print(f"statistiques descriptives pour {tickers_names[ticker]}:")
    print(clean_df_dict[ticker].describe())
    print("\n")

    plt.figure(figsize=(12, 8))
    
    plt.plot(clean_df_dict[ticker].index, clean_df_dict[ticker]['Close'], label=ticker)
    plt.title(f'Cours de {tickers_names[ticker]} depuis 2006')
    plt.xlabel('Date')
    plt.ylabel('Prix')
    plt.legend( loc='upper left')
    plt.show()





    
    



```

```python
# Mise en place de la stratégie A

def add_sma(df, short_window, long_window):
    df['SMA_short'] = df['Close'].rolling(window=short_window, min_periods=10, center=False).mean()
    df['SMA_long'] = df['Close'].rolling(window=long_window, min_periods=10, center=False).mean()
    return df
short_window = 50
long_window = 200

for ticker in tickers:
    clean_df_dict[ticker] = add_sma(clean_df_dict[ticker], short_window, long_window)

    plt.figure(figsize=(15, 10))
    plt.plot(clean_df_dict[ticker].index, clean_df_dict[ticker]['Close'], label=ticker)
    plt.plot(clean_df_dict[ticker].index, clean_df_dict[ticker]['SMA_short'], label=f'SMA {short_window} jours')
    plt.plot(clean_df_dict[ticker].index, clean_df_dict[ticker]['SMA_long'], label=f'SMA {long_window} jours')
    plt.title(f'Cours de {tickers_names[ticker]}') 
    plt.xlabel('Date')
    plt.ylabel('Prix')
    plt.legend(loc='upper left')
    plt.show()

######## Fonction pour générer les signaux




def generate_signals(data, short_window, long_window):
    data = add_sma(data, short_window, long_window)
    data['Signal'] = 0
    data['Signal'][short_window:] = np.where(data['SMA_short'][short_window:] > data['SMA_long'][short_window:], 1, 0)
    data['Position'] = data['Signal'].diff()
    return data


plt.figure(figsize=(10, 9))

for ticker in tickers:

    clean_df_dict[ticker] = generate_signals(clean_df_dict[ticker],  short_window, long_window )
    
    plt.plot(clean_df_dict[ticker]['Close'], label=ticker)
    plt.plot(clean_df_dict[ticker]['SMA_short'], label=f'SMA {short_window} jours')
    plt.plot(clean_df_dict[ticker]['SMA_long'], label=f'SMA {long_window} jours')
    plt.plot(clean_df_dict[ticker][clean_df_dict[ticker]['Position'] == 1].index,
             clean_df_dict[ticker]['SMA_short'][clean_df_dict[ticker]['Position'] == 1],
             '^', markersize=10, color='g', lw=0, label='Achat')
    plt.plot(clean_df_dict[ticker][clean_df_dict[ticker]['Position'] == -1].index,
             clean_df_dict[ticker]['SMA_short'][clean_df_dict[ticker]['Position'] == -1],
             'v', markersize=10, color='r', lw=0, label='Vente')
    plt.title(f'Prix de clôture et SMA avec Signaux de Trading pour {tickers_names[ticker]}')
    plt.xlabel('Date')
    plt.ylabel('Prix')
    plt.legend()
    plt.show()
```

```python
# Fonction de backtesting

def backtest(data, initial_capital=10000):
    positions = pd.DataFrame(index=data.index).fillna(0.0)
    positions['Position'] = 100 * data['Signal']   # Acheter 100 actions ou équivalent
    portfolio = positions.multiply(data['Close'], axis=0)
    pos_diff = positions.diff()
    portfolio['Holdings'] = positions.multiply(data['Close'], axis=0)
    portfolio['Cash'] = initial_capital - (pos_diff.multiply(data['Close'], axis=0)).cumsum()
    portfolio['Total'] = portfolio['Cash'] + portfolio['Holdings']
    portfolio['Returns'] = portfolio['Total'].pct_change()
    return portfolio

portfolios = {}
for ticker in tickers:
    portfolios[ticker] = backtest(clean_df_dict[ticker])


    plt.figure(figsize=(15, 8))
    plt.plot(portfolios[ticker]['Total'], label='Valeur du Portefeuille')
    plt.title(f'Backtesting de la Stratégie de Trading pour {tickers_names[ticker]}')
    plt.xlabel('Date')
    plt.ylabel('Valeur du Portefeuille')
    plt.legend()
    plt.show()
```

```python
# Fonction d'optimisation

def optimize_strategy(data, short_window_range, long_window_range):
    best_performance = -np.inf
    best_short_window = None
    best_long_window = None
    for short_window in short_window_range:
        for long_window in long_window_range:
            if short_window >= long_window:
                continue
            signals = generate_signals(data.copy(), short_window, long_window)  # Fix: Pass the additional arguments
            portfolio = backtest(signals)
            performance = portfolio['Total'].iloc[-1]

            if performance > best_performance:
                best_performance = performance
                best_short_window = short_window
                best_long_window = long_window
    return best_short_window, best_long_window, best_performance

short_window_range = range(20, 90, 5)
long_window_range = range(100, 200, 10)

best_params = {}
for ticker in tickers:
   short_window, long_window, performance = optimize_strategy(clean_df_dict[ticker], short_window_range, long_window_range)
   
   best_params[ticker] = (short_window, long_window, performance)
   print(f"Ticker: {tickers_names[ticker]}, Short Window: {short_window}, Long Window: {long_window}, Performance: {performance}")

# Afficher les paramètres optimaux

print("\nParamètres optimaux pour chaque ticker:")
for ticker in tickers:
    short_window, long_window, performance = best_params[ticker]
    print(f"{tickers_names[ticker]}: Short Window = {short_window}, Long Window = {long_window}, Performance = {performance}")
```

```python
evaluated_portfolios = {}
for ticker in tickers:
    short_window, long_window, _ = best_params[ticker]
    signals = generate_signals(clean_df_dict[ticker], short_window, long_window)
    evaluated_portfolios[ticker] = backtest(signals)
```
